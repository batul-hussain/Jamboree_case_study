```python
#Feature Engineering

# Feature extraction i.e removing low level feature unsuitable for model learning
df.drop(columns=["Serial No."],inplace=True)

#Scaling and Normalization
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df_scaled = ss.fit_transform(df)
df = pd.DataFrame(df_scaled,columns = df.columns)
```

```python
df.head()
```

Data Preparation for Modeling

```python
#Putting feature variable to X and response variable to y
X = df[df.columns.drop('Chance of Admit')]
y = df['Chance of Admit']
```

Splitting Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.7)
```


# **MODEL BUILDING** 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Lets print intercept
print(lr.intercept_)
```

```python
#Lets print coefficients
coef_df = pd.DataFrame(lr.coef_,X_test.columns,columns = ["Coefficients"])
coef_df
```

From the coefficient values we cobserve that SOP, University Rating and Research doesnt have significant impact on Chance of Admission in the college. 

```python
#predictions
y_pred = lr.predict(X_test)
```

```python
#Actual vs Predicted
import matplotlib.pyplot as plt
c = [i for i in range(1,len(y_test)+1)]
fig = plt.figure()
plt.plot(c,y_test,color = "blue",linewidth=2.5,linestyle="-")
plt.plot(c,y_pred,color = "red",linewidth=2.5,linestyle="-")
fig.suptitle("Actual and Predicted",fontsize = 15)
plt.xlabel("Index",fontsize = 16)
plt.ylabel("Chance of Admit",fontsize = 16)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Y Test")
plt.ylabel("Predicted Y")
```

ERROR TERMS

```python
#Error Terms
c = [i for i in range(1,len(y_test)+1)]
fig = plt.figure()
plt.plot(c,y_test-y_pred, color = "green", linewidth = 2,linestyle="-")
fig.suptitle("Error Terms",fontsize = 15)
plt.xlabel("Index",fontsize = 16)
plt.ylabel("y_test-y_pred",fontsize = 16)
```

```python
sns.histplot(y_test-y_pred,kde = True)
#Residuals are normally distributed
```

```python
sns.scatterplot(y_pred,y_test-y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test,y_pred)
r_squared = r2_score(y_test,y_pred)
adjr2_sq = 1 - ( 1-r_squared) * ( len(y_test) - 1 ) / ( len(y_test) - X_test.shape[1] - 1 )
```

```python
print("mean_squared_error :",mse)
print("r_squared value :",r_squared)
print("adjusted r_squared value :",adjr2_sq)
```

R_squared and adjusted r_squared values measures goodness to fit of the model. R-Squared above 0.7 would generally be seen as showing a high level of correlation however it depends on the business case and problem statement 

**Checking for p value using statsmodels**

```python
import statsmodels.api as sm
X_train_sm = X_train
X_train_sm = sm.add_constant(X_train_sm)
ml = sm.OLS(y_train,X_train_sm).fit()

print(ml.summary())
```

==> Null hypothesis: A coefficient equals zero.

==> Alternate Hypothesis: A coefficient is non-zero.

Since p-value is less than 0.05 for GRE Score, TOEFL Score,LOR,CGPA and Research, it indicates a strong evidence against null hypothesis which states that the corresponding independent variable has an effect on the dependent variable

# **Model performance Evaluation**

**Mean Absolute Error(MAE)** is similar to **Mean Square Error(MSE)**.MSE gives larger penalization to big prediction error by square it while MAE treats all errors the same.

**R Square** is a good measure to determine how well the model fits the dependent variables. However, it does not take into consideration of overfitting problem.That is why **Adjusted R Square** is introduced because it will penalize additional independent variables added to the model and adjust the metric to prevent overfitting issues.

```python
from sklearn.metrics import mean_absolute_error
mae_new = mean_absolute_error(y_test,y_pred_new)
mse_new = mean_squared_error(y_test,y_pred_new)
r_squared_new = r2_score(y_test,y_pred_new)
adjr2_sq_new = 1 - ( 1-r_squared_new) * ( len(y_test) - 1 ) / ( len(y_test) - X_test_new.shape[1] - 1 )
```

```python
print("mean_absolute_error :",mae_new)
print("mean_squared_error :",mse_new)
print("r_squared value :",r_squared_new)
print("adjusted r_squared value :",adjr2_sq_new)
```

Removing the outliers and than training the model again might improve R_squared and Adjusted R_squared values
